## Agrupaci√≥n de vac√≠os por $R_v$, $z$, $\rho_1$, $\rho_2$


```python
def lenscat_load(Rv_min, Rv_max, z_min, z_max, rho1_min, rho1_max, rho2_min, rho2_max,
                 flag=2, lensname = '/home/franco/FAMAF/Lensing/cats/MICE/voids_MICE.dat'):

    L = pd.DataFrame(np.loadtxt(lensname), columns=['id', 'R_v', 'alpha', 'delta', 'z', 'xv', 'yv', 'zv', 'rho1', 'rho2', 'logp', 'flag'])
    m_rv = (L.R_v >= Rv_min) & (L.R_v <= Rv_max)
    m_z = (L.z >= z_min) & (L.z <= z_max)
    m_rho = (L.rho1 >= rho1_min) & (L.rho1 <= rho1_max) & (L.rho2 >= rho2_min) & (L.rho2 <= rho2_max)
    m_flag = (L.flag >= flag)
    mask = m_rv & m_z & m_rho & m_flag

    L = L.astype({'id':int})

    return L[mask]
```

```python
def profiles(L):
    profs = {}
    for idd in L.id:
        try:
            profs[idd] = read_profileSCVF(idd)
        except:
            print(f'error in {idd}')
            continue

    DeltaDiff_tot = np.zeros(100)
    DeltaCum_tot  = np.zeros(100)
    for ii in profs.keys():
        DeltaDiff_tot += profs[ii].DeltaDiff + 1
        DeltaCum_tot += profs[ii].DeltaCum + 1

    nvoids = L.shape[0]
    print(nvoids)

    DeltaDiff_tot /= nvoids
    DeltaCum_tot /= nvoids

    return DeltaDiff_tot-1 , DeltaCum_tot-1
```

```python
def plot_profs(profs):
    r = np.linspace(0.05,5.0,100)
    plt.figure()
    plt.plot(r, profs[0])
    plt.plot(r, profs[1])
    plt.axhline(0,ls='--',c='k',alpha=0.5)
    plt.show()
```

```python
Rv_min, Rv_max, z_min, z_max, rho1_min, rho1_max, rho2_min, rho2_max = 10.0, 15.0, 0.2, 0.3, -1.0, -0.9, -1.0, 100.0
args = (Rv_min, Rv_max, z_min, z_max, rho1_min, rho1_max, rho2_min, rho2_max)

def V(L, vvv=False):
    profs = {}
    for idd in L.id:
        try:
            profs[idd] = read_profileSCVF(idd)
        except:
            if vvv:
                print(f'error in {idd}')
            continue
    return profs

voids = V(lenscat_load(*args))

# print(voids[30074].DeltaDiff)
r = np.linspace(0.05, 5, 100)
# plt.figure(figsize=(16,9))
D = np.zeros((100,len(voids)))
D_c = np.zeros((100, len(voids)))
for i,v in enumerate(voids.values()):
    # plt.plot(r, v.DeltaDiff)
    # plt.plot(r, v.DeltaCum)
    D[:,i] += v.DeltaDiff
    D_c[:,i] += v.DeltaCum
```
